```python
from selenium import webdriver
from selenium.webdriver.edge.service import Service
from webdriver_manager.microsoft import EdgeChromiumDriverManager
from selenium.webdriver.common.by import By
from selenium.webdriver import Keys, ActionChains
from random import randint
from time import sleep
import pyautogui


cordenadas_login = [[280,605], [280,640], [280,675], [280,710]]# 35 de diferença
conta1 = r'E:\Repos-Github\Pesquisa-Microsoft-Rewards\conta-1.png'
conta2 = r'E:\Repos-Github\Pesquisa-Microsoft-Rewards\conta-2.png'
conta_teste = r'E:\Repos-Github\Pesquisa-Microsoft-Rewards\conta-teste.png'
lista_contas = [conta1, conta2]
#lista_contas = [conta_teste] ########## teste
letras = 'askdjalskjdkasjdlkajskljfrhhbcamzjznhcas'
inspecionar = r'E:\Repos-Github\Pesquisa-Microsoft-Rewards\inspecionar.png'
mobile = r'E:\Repos-Github\Pesquisa-Microsoft-Rewards\mobile.png'
wait_inspecionar = r'E:\Repos-Github\Pesquisa-Microsoft-Rewards\wait-inspecionar.png'
jogar = r'E:\Repos-Github\Pesquisa-Microsoft-Rewards\jogar.png'
jogar2 = r'E:\Repos-Github\Pesquisa-Microsoft-Rewards\jogar-2.png'
fechar_inspec = r'E:\Repos-Github\Pesquisa-Microsoft-Rewards\fechar-inspec.png'
caminho_classe = 'StreamGate-module__primaryButton___Wyg1n Button-module__basicBorderRadius___pIIUL Button-module__defaultBase___+Z-eV Button-module__buttonBase___LuYUw Button-module__callToAction___x1-O7 Button-module__textNoUnderline___5pGM- Button-module__typePrimary___AR+oQ Button-module__sizeMedium___WZn2T Button-module__overlayModeSolid___2BVUO'

# sincronizando a versão do navegador com o webdriver e iniciando o mesmo
serv = Service(EdgeChromiumDriverManager().install())
```

```python
num_conta = 0

for conta in lista_contas:
    num_desktop = 0
    num_mobile = 0
    num_aleatorio = randint(0,999999999999999999)
    # abrindo navegador
    navegador = webdriver.Edge(service=serv)
    # fazendo login
    navegador.get('edge://settings/profiles')
    # esperando e clicando para abrir o menu
    while len(navegador.find_elements(By.CLASS_NAME, 'c01437')) == 0:
        sleep(0.5)
    navegador.find_element(By.CLASS_NAME, 'c01437').click()
    # esperando abrir o menu
    while not pyautogui.locateCenterOnScreen(conta):
        sleep(0.5)
    # clicando no email
    pyautogui.click(conta, duration=0.5)
    # clicando no entrar
    pyautogui.click(500, 535, duration=0.5)
    num_conta += 1
    print(f'Conta: {num_conta}')
    sleep(1.5)
    # entrando no site rewards
    navegador.get('https://rewards.bing.com/?form=dash_2')
    sleep(2)
    # verificando se tem ponto pra resgatar
    quantidade = navegador.find_elements(By.CLASS_NAME, 'mee-icon-AddMedium')
    print(f'Quantidade banners: {len(quantidade)}')
    # clicando nos cards pra pegar os pontos
    if len(quantidade) > 0:
        for elemento in quantidade:
            ActionChains(navegador).key_down(Keys.LEFT_CONTROL).click(elemento).key_up(Keys.LEFT_CONTROL).perform()
    # entrando no site para fazer as pesquisas
    navegador.get('https://www.bing.com/?FORM=Z9FD1')
    # esperando o elemento onde pesquisa carregar
    while len(navegador.find_elements(By.ID, 'sb_form_q')) == 0:
        sleep(0.5)
    sleep(2)
    # tirando popout
    if len(navegador.find_elements(By.ID, 'bnp_btn_reject')) > 0:
        navegador.find_element(By.ID, 'bnp_btn_reject').click()
        sleep(0.5)
    # pesquisa desktop
    navegador.find_element(By.ID, 'sb_form_q').send_keys(num_aleatorio)
    for letra in letras:
        navegador.find_element(By.ID, 'sb_form_q').send_keys(letra)
        navegador.find_element(By.ID, 'sb_form_q').send_keys(Keys.ENTER)
        num_desktop += 1
        if num_desktop == 38: ######### diminuir o numero no teste
            break
        sleep(0.5)
    print('Desktop: ok')
    sleep(1)
    while pyautogui.locateOnScreen(wait_inspecionar) == None:
        sleep(1)
    # entrando no mobile
    pyautogui.rightClick(wait_inspecionar, duration=0.5)
    pyautogui.click(inspecionar, duration=0.5)
    while pyautogui.locateOnScreen(mobile) == None:
        sleep(0.5)
    pyautogui.click(mobile, duration=0.5)
    # pesquisa mobile
    for letra in letras:
        navegador.find_element(By.ID, 'sb_form_q').send_keys(letra)
        navegador.find_element(By.ID, 'sb_form_q').send_keys(Keys.ENTER)
        num_mobile += 1
        if num_mobile == 24: ######### diminuir o numero no teste
            break
        sleep(0.5)
    print('Mobile: ok')
    while pyautogui.locateOnScreen(fechar_inspec) == None:
        sleep(1)
    pyautogui.click(fechar_inspec, duration=0.5)
    # entrando na pagina fortnite e colocando para jogar
    navegador.get('https://www.xbox.com/pt-BR/play/games/fortnite/BT5P2X999VH2')
    # esperando logar na conta microsoft
    while pyautogui.locateOnScreen(jogar) == None:
        sleep(0.5)
    # clicando em jogar
    navegador.find_element(By.CLASS_NAME, 'BaseItem-module__overlay___1SkWh').click()
    sleep(1)
    # verificando se o overlay aparece
    if pyautogui.locateOnScreen(jogar2) is not None:
        pyautogui.click(jogar2, duration=0.5)
    sleep(2)
    # saindo da tela cheia
    pyautogui.press('esc')
    print('Fortnite: ok')
    print('-'*20)

    # break para usar no teste ###############
    #break
```
